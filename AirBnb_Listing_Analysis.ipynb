```python
import pandas as pd
import numpy as np
```

```python
listings = pd.read_csv('/Users/pankit/Downloads/Listings.csv', encoding="ISO-8859-1", low_memory=False)
```

```python
listings.head()
```

```python
listings.info()
```

```python
listings["host_since"] = pd.to_datetime(listings["host_since"])

listings.info()
```

```python
listings_paris = listings.query("city == 'Paris'")

listings_paris.info()
```

```python
#We have 64690 entries with city as Paris.
#Now we take only the columns which are required for us

listings_paris = (listings.query("city == 'Paris'").loc[:,["host_since", "neighbourhood", "city", "accommodates"
                                                         , "price"]])
listings_paris.info()
```

```python
#Checking missing values in our data

listings_paris.isna().sum()
```

```python
listings_paris.describe()
```

```python
listings_paris.query("accommodates == 0").count()
```

```python
#From the above we can conclude, if accommodates is 0 then price will be 0
```

```python
listings_paris_neighbourhood = (listings_paris
                                .groupby("neighbourhood")
                                .agg({"price":"mean"})
                                .sort_values("price"))

listings_paris_neighbourhood.tail()
```

```python
listings_paris_accommodates = (listings_paris
                              .query("neighbourhood == 'Elysee'")
                              .groupby("accommodates")
                              .agg({"price" : "mean"})
                              .sort_values("price"))

listings_paris_neighbourhood.head()
```

```python
listings_paris_over_time = (listings_paris
                            .set_index("host_since")
                            .resample("Y")
                            .agg({"neighbourhood" : "count",
                                  "price" : "mean"})
                           )

listings_paris_over_time.head()
```

```python
listings_paris_neighbourhood.plot.barh(title = "Average Listing Price",
                                       legend = None)
```

```python
listings_paris_accommodates.plot.barh(title = "Average Listing Price by Accommodation",
                                       legend = None)
```

```python
listings_paris_over_time["neighbourhood"].plot(
      ylabel = "New Hosts",
      title = "New AirBBnb Hosts in Paris Over Time")
```

```python
listings_paris_over_time["price"].plot(
      ylabel = "Average Price (Euros)",
      title = "New AirBBnb Price in Paris Over Time")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(listings_paris_over_time.index,
        listings_paris_over_time["neighbourhood"],
        label = "New Hosts",
        c = "pink")



ax.set_ylabel("New Hosts")

ax2 = ax.twinx()

ax2.plot(listings_paris_over_time.index,
        listings_paris_over_time["price"],
        label = "Average Price")

ax2.set_ylim(0)

ax2.set_ylabel("Average Price")

ax.set_title("2015 Regulations Lead to Less Hosts but Higher Price")
```
